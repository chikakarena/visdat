Import Libary

```python
import matplotlib.pyplot as plt
```

Dataset

```python
# buat data sample
months = ['Jan', 'Feb','Mar','Apr','Mei', 'Jun', 'Jul', 'Aug','Sep','Oct', 'Nov', 'Des']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

Single line Plot

```python
#line plot 1 garis, style grafik default
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid (True)
plt.show()
```

Multiple Line plot and Customizations

```python
#Kustomisasi line plot (line 1 - product A)
plt.plot(months, product_A_sales, label='Product A', linestyle="--",
marker='o', color='blue')
#Kustomisasi line plot (line 2 - product B)
plt.plot(months, product_A_sales, label='Product B', linestyle="-",
marker='x', color='red')

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid (True)
plt.show()
```

## Jenis Garis untuk Menunjukan Tren

```python
import matplotlib.pyplot as plt

# buat data sample
months = ['Jan', 'Feb','Mar','Apr','Mei', 'Jun', 'Jul', 'Aug','Sep','Oct', 'Nov', 'Des']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]

# Data sample tambahan
product_C_sales = [18,22,25,28,32,38,42,45,48,52,56,60]
product_D_sales = [7,9,11,13,16,18,20,23,25,28,30,33]

# plot product A
plt.plot(months, product_A_sales, label='Product A', linestyle='--',color='Blue')

# Kustomisasi Line plot (line 2 - Product B)
plt.plot(months, product_B_sales, label='Product B', linestyle='-', color='red')

# Kustomisasi Line plot (line 2 - Product C)
plt.plot(months, product_C_sales, label='Product C', linestyle=':', color='green')

# Kustomisasi Line plot (line 2 - Product D)
plt.plot(months, product_D_sales, label='Product D', linestyle='-.', color='purple')

# Tambahkan judul dan label
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

Subplot pada Line Chart

```python
# subplot, 2 baris 1 kolom, panjang 10 inch, tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk product C
axs[0].plot(months, product_C_sales, label='Product C', linestyle='-', color='green', marker='d')
axs[0].set_title('Product Sales C Over Months')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Sales')
axs[0].legend()
axs[0].grid(True)

# plot pertama untuk product D
axs[1].plot(months, product_D_sales, label='Product D', linestyle='-', color='purple', marker='s')
axs[1].set_title('Product Sales D Over Months')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Sales')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
